## Part 6: Linear Algebra & Matrix Operations (2 Questions)

#### 1) Compute the determinant and inverse of a square matrix of your choice.

```python
arr = np.array([[2, 1, 3],
                [1, 0, 2],
                [4, 1, 8]])

det = np.linalg.det(arr)
print("Determinant:", det)

inv = np.linalg.inv(arr)
print("Inverse:\n", inv)
```
